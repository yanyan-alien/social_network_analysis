```python
import networkx as nx
import pandas as pd
import numpy as np
import pickle
```

```python
G = nx.Graph()
G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('D', 'C')
G.add_edge('B', 'D')
G.add_edge('E', 'D')
G.add_edge('E', 'C')
G.add_edge('G', 'D')
G.add_edge('E', 'G')
G.add_edge('F', 'G')
```

```python
draw_pos=nx.spring_layout(G, seed=2)
nx.draw(G, with_labels=True)
```
